```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv(r"C:\Users\preethi\Documents\Sample Datasets\Unemployment in India.csv")
df
```

```python
# Data Cleaning - Fix Structural Errors and Handling Misssing Values
df.isna().sum() # Total Null values are 28 in all column fields
```

```python
# Removing Null Values
df=df.dropna()
df # Deleted 28 empty fields
```

```python
df.columns=[i.strip() for i in df.columns] # Removing White Spaces in Column Names
df.columns
```

```python
# Fixing Structural error (Converting Date column Field as Datetime Column Field)
df.loc[:,'Date']=pd.to_datetime(df['Date'])
df
```

```python
df.loc[:,'Frequency']=[i.strip() for i in df['Frequency'].values]
df
```

```python
# Selecting Region == AndhraPradesh
df_AP=df.loc[df['Region']=='Andhra Pradesh']
df_AP
```

```python
df_AP=df_AP.groupby(['Date','Area'],as_index=False).agg({c,'Estimated Employed':'sum',
                                   'Estimated Labour Participation Rate (%)':'sum'})
```

```python
x=df_AP['Date'].apply(lambda x:x.strftime("%Y-%m")).unique()
x
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.color_palette('magma',as_cmap=True)
ax=sns.barplot(df_AP,x='Date',y='Estimated Unemployment Rate (%)',hue='Area',palette='rocket')
ax.set_xticklabels(x)
plt.title("Unemployment Rate in AndhraPradesh State for the period May 2019 to June 2020")
plt.show()
```

```python
df.groupby(['Date','Area'],as_index=False).agg({'Estimated Unemployment Rate (%)':'max'})
```

```python
result={}
for i in df['Date'].unique():
    x=df.loc[(df['Date']==i) & (df['Area']=='Rural'),['Region','Estimated Unemployment Rate (%)']].sort_values(by='Estimated Unemployment Rate (%)',ascending=False).head(1).values[0]
    result[i]=x
```

```python
# State with High Unemployment Rate for Rural Area
rural_highest=pd.DataFrame(result).T.rename(columns={0:'State',1:'Unemployment Rate(%)'})
rural_highest
```

```python
urban={}
for i in df['Date'].unique():
    x=df.loc[(df['Date']==i) & (df['Area']=='Urban'),['Region','Estimated Unemployment Rate (%)']].sort_values(by='Estimated Unemployment Rate (%)',ascending=False).head(1).values[0]
    urban[i]=x
    
urban_highest=pd.DataFrame(urban).T.rename(columns={0:'State',1:'Unemployment Rate(%)'})
urban_highest
```

```python
urban_low={}
for i in df['Date'].unique():
    x=df.loc[(df['Date']==i) & (df['Area']=='Urban') & (df['Estimated Unemployment Rate (%)']>0),['Region','Estimated Unemployment Rate (%)']].sort_values(by='Estimated Unemployment Rate (%)',ascending=True).head(1).values[0]
    urban_low[i]=x
    
urban_lowest=pd.DataFrame(urban_low).T.rename(columns={0:'State',1:'Unemployment Rate(%)'})
urban_lowest
```

```python
rural_low={}
for i in df['Date'].unique():
    x=df.loc[(df['Date']==i) & (df['Area']=='Rural') & (df['Estimated Unemployment Rate (%)']>0),['Region','Estimated Unemployment Rate (%)']].sort_values(by='Estimated Unemployment Rate (%)',ascending=True).head(1).values[0]
    rural_low[i]=x
    
rural_lowest=pd.DataFrame(rural_low).T.rename(columns={0:'State',1:'Unemployment Rate(%)'})
rural_lowest
```
